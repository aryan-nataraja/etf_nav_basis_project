```python
import pandas as pd
import numpy as np

df = pd.read_csv('../data/processed/tlt_nav_basis.csv', parse_dates=['DATE']).set_index('DATE')
df = df[['etf_close', 'nav_proxy']].dropna()
```

Check for stationarity

```python
from statsmodels.tsa.stattools import adfuller

for col in ['etf_close', 'nav_proxy']:
    result = adfuller(np.log(df[col]))
    print(f'{col}: p-value = {result[1]:.4f}')
```

Test for cointegration

```python
from statsmodels.tsa.stattools import coint

score, pvalue, _ = coint(np.log(df['etf_close']), np.log(df['nav_proxy']))
print(f'Cointegration test p-value: {pvalue:.4f}')
```

Given the generation of our proxy NAV was only approximate - it has not captured all the features of movement. As a result, p-value > 0.9917 indicates that the NAV and ETF are not cointegrated. Further formal cointegration will be skipped
